```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
combined_data.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
mean_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdv_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
summary_df = pd.DataFrame({"Mean": mean_summ, "Median": median_summ, "Variance": variance_summ, "Standard Deviation": stdv_summ, 
                          "SEM": sem_summ})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
summary_df = round(pd.DataFrame({'Tumor Volume (mm3) Mean': mean_summ, 'Median': median_summ, 'Variance': variance_summ, 'Standard Deviation': stdv_summ, 
                        'Std Err': sem_summ}),2)
summary_df.head()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_great_df = combined_data.groupby("Mouse ID")['Timepoint'].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
sort_df = combined_data.merge(last_great_df, on=['Mouse ID','Timepoint'], how='inner')
sort_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments_list =['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumor_data = []

# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
      # Locate the rows which contain mice on each drug and get the tumor volumes
for drug_name in treatments_list:
    last_volume = sort_df.loc[sort_df['Drug Regimen']== drug_name, 'Tumor Volume (mm3)']
    tumor_data.append(last_volume)

    # add subset 
    # Determine outliers using upper and lower bounds
quartiles = last_volume.quantile ([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles [0.75]
iqr =upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
pot_outliers = last_volume.loc[(last_volume < lower_bound) | (last_volume > upper_bound)]
print(f"The {drug_name}'s potential outliers : {pot_outliers}")    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_data, labels=treatments_list, flierprops={'markerfacecolor':'b', 'markersize':7})
plt.title("The Final Tumor Volume - Mouse")
plt.show()
```
