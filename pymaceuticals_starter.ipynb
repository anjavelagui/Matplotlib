## Observations and Insights 

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mice_tot = combined_data['Mouse ID'].count()
mice_tot
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_byID = combined_data[combined_data.duplicated(["Mouse ID", 'Timepoint'])]
duplicate_byID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
all_duplicate_byID = combined_data[combined_data.duplicated(["Mouse ID"])]
all_duplicate_byID
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouseID_df = combined_data.drop_duplicates(["Mouse ID"])
clean_mouseID_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouseID_df = combined_data['Mouse ID'].count()
clean_mouseID_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
mean_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdv_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem_summ = combined_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
summary_df = pd.DataFrame({"Mean": mean_summ, "Median": median_summ, "Variance": variance_summ, "Standard Deviation": stdv_summ, 
                          "SEM": sem_summ})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
summary_df = round(pd.DataFrame({'Tumor Volume (mm3) Mean': mean_summ, 'Median': median_summ, 'Variance': variance_summ, 'Standard Deviation': stdv_summ, 
                        'Std Err': sem_summ}),2)
summary_df.head()
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.
drug_count = combined_data["Drug Regimen"].value_counts()
drug_count.plot(kind='bar', xlabel='Drug Regimen', ylabel='Count of Measurements', title = 'Total Measurements per Drug Regimen')
```

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pyplot.
plt.bar(combined_data['Drug Regimen'].unique(), combined_data['Drug Regimen'].value_counts())
plt.xticks(rotation=45)
plt.margins(0.1)
plt.title('Total Measurements per Drug Regimen')
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
combined_data['Sex'].value_counts().plot(kind='pie', y='Mouse ID', title = 'Male vs Female Distribution', autopct='%1.1f%%')
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plt.pie(combined_data['Sex'].value_counts(), labels=combined_data['Sex'].unique(), explode=(0.1,0), autopct='%1.1f%%')
plt.title('Male vs Female Distribution')
```
